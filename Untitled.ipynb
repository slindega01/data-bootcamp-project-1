```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
from scipy.stats import linregress
```

```python
crime_data = pd.read_csv('../../datasets/data_2010/crime_2010_county_totals.csv')
crime_data = crime_data.set_index('County')
crime_data.index =  crime_data.index.astype(str)+' County'
crime_data.rename(index={'St Lawrence': 'St. Lawrence'},inplace=True)
crime_data.head()
```

```python
population_data = pd.read_csv('../../datasets/data_2010/2010_racial_diversity_county.csv')
population_data.columns = population_data.iloc[0]
del population_data['Estimate Total Two or more races Two races including Some other race']
del population_data['Estimate Total Two or more races Two races excluding Some other race, and three or more races']
population_data["County"] = population_data["Geographic Area Name"].str.split(",").str[0]
del population_data["Geographic Area Name"]
del population_data["id"]
population_data = population_data.drop([0])
population_data = population_data.set_index('County')
population_data.head()
#population_data.count()
```

```python
merged_df = population_data.merge(crime_data,how='inner',on='County')
merged_df["Estimate Total"] = pd.to_numeric(merged_df["Estimate Total"])
merged_df["Estimate Total White alone"] = pd.to_numeric(merged_df["Estimate Total White alone"])
merged_df["Estimate Total Black or African American alone"] = pd.to_numeric(merged_df["Estimate Total Black or African American alone"])
merged_df["Estimate Total American Indian and Alaska Native alone"] = pd.to_numeric(merged_df["Estimate Total American Indian and Alaska Native alone"])
merged_df["Estimate Total Asian alone"] = pd.to_numeric(merged_df["Estimate Total Asian alone"])
merged_df["Estimate Total Native Hawaiian and Other Pacific Islander alone"] = pd.to_numeric(merged_df["Estimate Total Native Hawaiian and Other Pacific Islander alone"])

merged_df.head()
```

```python
total_2010 = merged_df.iloc[:, 0]
white_2010 = merged_df.iloc[:, 1]
black_2010 = merged_df.iloc[:, 2]
american_indian_2010 = merged_df.iloc[:, 3]
asian_2010 = merged_df.iloc[:, 4]
pacific_2010 = merged_df.iloc[:, 5]
other_2010 = merged_df.iloc[:, 6]
race2_2010 = merged_df.iloc[:, 7]
total_crime_2010 = merged_df.iloc[:, 9]
violent_crime_2010 = merged_df.iloc[:, 10]
property_crime_2010 = merged_df.iloc[:, 15]
```

```python
#regression lines
#White Population
slope1,int1,r1,p1,std1 = linregress(white_2010, total_crime_2010)
line1 = "y =" +str(round(slope1,2)) + "x +" + str(round(int1,2))
fit1 = slope1 *white_2010 + int1
r_squ1 = "R-squared" +str(round(r1**2,2))

#African American Population
slope2,int2,r2,p2,std2 = linregress(black_2010, total_crime_2010)
line2 = "y =" +str(round(slope2,2)) + "x +" + str(round(int2,2))
fit2 = slope2 *black_2010 + int2
r_squ2 = "R-squared" +str(round(r2**2,2))

#American Indian/Alaskan Native Population
slope3,int3,r3,p3,std3 = linregress(american_indian_2010, total_crime_2010)
line3 = "y =" +str(round(slope3,2)) + "x +" + str(round(int3,2))
fit3 = slope3 *american_indian_2010 + int3
r_squ3 = "R-squared" +str(round(r3**2,2))

#Asian Population
slope4,int4,r4,p4,std4 = linregress(asian_2010, total_crime_2010)
line4 = "y =" +str(round(slope4,2)) + "x +" + str(round(int4,2))
fit4 = slope4 *asian_2010 + int4
r_squ4 = "R-squared" +str(round(r4**2,2))

#Hawaiian/Pacific Islander Population
slope5,int5,r5,p5,std5 = linregress(pacific_2010, total_crime_2010)
line5 = "y =" +str(round(slope5,2)) + "x +" + str(round(int5,2))
fit5 = slope5 *pacific_2010 + int5
r_squ5 = "R-squared" +str(round(r5**2,2))

#Total Population
slope6,int6,r6,p6,std6 = linregress(total_2010, total_crime_2010)
line6 = "y =" +str(round(slope6,2)) + "x +" + str(round(int6,2))
fit6 = slope6 *total_2010 + int6
r_squ6 = "R-squared" +str(round(r6**2,2))

#graphs
fig, (graph1,graph2,graph3,graph4,graph5,graph6) = plt.subplots(6,figsize=(10,15),sharex=False,sharey=True)
fig.tight_layout()
fig.suptitle("Total Crime Instances vs. Population Demographics", fontsize = 20, y = 1.05)

graph1.scatter(white_2010, total_crime_2010, color = "Purple")
graph1.set_xlabel("White Population")
graph1.set_ylabel("Total Crime")
graph1.plot(white_2010,fit1,"Purple")
graph1.set_xlim(0,max(white_2010)+8000)
graph1.set_ylim(0,max(total_crime_2010)+8000)
graph1.annotate(line1,(max(white_2010)*.7,10000),color="Purple",clip_on=True)
graph1.annotate(r_squ1,(max(white_2010)*.7,1000),color="Purple",clip_on=True)

graph2.scatter(black_2010, total_crime_2010, color = "Blue")
graph2.set_xlabel("African American Population")
graph2.set_ylabel("Total Crime")
graph2.plot(black_2010,fit2,"Blue")
graph2.annotate(line2,(max(white_2010)*.7,10000),color="Blue",clip_on=True,ha='center')
graph2.annotate(r_squ2,(max(white_2010)*.7,1000),color="Blue",clip_on=True,ha='center')
graph2.set_xlim(0,max(white_2010)+8000)
graph2.set_ylim(0,max(total_crime_2010)+8000)

graph3.scatter(american_indian_2010, total_crime_2010, color="Black")
graph3.set_xlabel("American Indian/Alaskan Native Population")
graph3.set_ylabel("Total Crime")
graph3.plot(american_indian_2010, fit3 ,"Black")
graph3.annotate(line3,(max(white_2010)*.7,10000),color="Black",clip_on=True)
graph3.annotate(r_squ3,(max(white_2010)*.7,1000),color="Black",clip_on=True)
graph3.set_xlim(0,max(white_2010)+8000)
graph3.set_ylim(0,max(total_crime_2010)+8000)

graph4.scatter(asian_2010, total_crime_2010, color = 'Red')  
graph4.set_xlabel("Asian Population")
graph4.set_ylabel("Total Crime")
graph4.plot(asian_2010,fit4,"Red")
graph4.annotate(line4,(max(white_2010)*.7,10000),color="Red",clip_on=True)
graph4.annotate(r_squ4,(max(white_2010)*.7,1000),color="Red",clip_on=True)
graph4.set_xlim(0,max(white_2010)+8000)
graph4.set_ylim(0,max(total_crime_2010)+8000)

graph5.scatter(pacific_2010, total_crime_2010, color = "darkorange") 
graph5.set_xlabel("Hawaiian/Pacific Islander Population")
graph5.set_ylabel("Total Crime")
graph5.plot(pacific_2010,fit5,"darkorange")
graph5.annotate(line5,(max(white_2010)*.7,10000),color="darkorange",clip_on=True)
graph5.annotate(r_squ5,(max(white_2010)*.7,1000),color="darkorange",clip_on=True)
graph5.set_xlim(0,max(white_2010)+8000)
graph5.set_ylim(0,max(total_crime_2010)+8000)

graph6.scatter(total_2010,total_crime_2010, color = "Green")
graph6.set_xlabel("Total Population")
graph6.set_ylabel("Total Crime")
graph6.plot(total_2010,fit6,"Green")
graph6.annotate(line6,(max(white_2010)*.7,10000),color="Green",clip_on=True)
graph6.annotate(r_squ6,(max(white_2010)*.7,1000),color="Green",clip_on=True)
graph6.set_xlim(0,max(white_2010)+8000)
graph6.set_ylim(0,max(total_crime_2010)+8000)
```
